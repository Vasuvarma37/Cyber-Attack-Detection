```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv"
)

print(df.shape)
```

```python
df.columns = df.columns.str.strip()
```

```python
df.replace(
    [np.inf, -np.inf],
    np.nan,
    inplace=True
)
```

```python
print(df.isnull().sum().sum())
```

```python
df.dropna(inplace=True)

df.drop_duplicates(inplace=True)

print(df.shape)
```

```python
df["Label"] = (
    df["Label"]
    .astype(str)
    .str.strip()
)

print(df["Label"].value_counts())
```

```python
print(df["Label"].isnull().sum())
```

```python
df.to_csv(
    "cleaned_cicids2017_v3.csv",
    index=False
)
```

```python
print(df.shape)

print(df["Label"].value_counts(dropna=False))
```
